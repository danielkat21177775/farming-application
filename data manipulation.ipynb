```r
library(ggplot2)


library(readxl)


library(dplyr)

library(openxlsx)


```

```r
blok_rye_dat = read_excel("D:\\Users\\User\\Documents\\BlokenRy.xlsx")

head(blok_rye_dat)
```

```r



## We want to see how many rows each blok of vineyards has:

rye_dat = as.data.frame(table(blok_rye_dat$Blok))

names(rye_dat) = c("Blok_nom", "Rye")

### Next we want to see the total amount of vakke each blok has.

vakke_dat <- aggregate(Fakke~Blok,blok_rye_dat,sum)

vak_ry_dat = cbind(rye_dat, vakke_dat[,2])

names(vak_ry_dat) = c("Blok_nom", "Rye" , "Vakke")


# print(vak_ry_dat)


# # Let's find out how much vines each blok has:

stokke_nom = rep(times = 26 , 7)

 stokke_nom[c(1, 20, 21)] = 4
 
mat = as.matrix(stokke_nom)




vak_ry_dat = as.matrix(vak_ry_dat)


vak_ry_dat = cbind(vak_ry_dat , mat)

colnames(vak_ry_dat) = c("Blok_nom", "Rye" , "Vakke" , "Vines_per_vak")

vak_ry_dat = as.data.frame(vak_ry_dat)

# Let's convert the columns of vak_ry_dat into numeric, except for the first column.



vak_ry_dat$Vakke = as.numeric(vak_ry_dat$Vakke)



vak_ry_dat$Rye = as.numeric(vak_ry_dat$Rye)


vak_ry_dat$Vines_per_vak = as.numeric(vak_ry_dat$Vines_per_vak)


## Let's create a column which indicates the number of vines for each block.


vak_ry_dat$Vines = vak_ry_dat$Vines_per_vak * vak_ry_dat$Rye * vak_ry_dat$Vakke



vak_ry_dat = vak_ry_dat[,!c(colnames(vak_ry_dat) == "Vines_per_vak")]
```

```r
# Write the vak_ry_dat dataset into excel.



write.xlsx(vak_ry_dat , "G:\\My Drive\\Daniel\\data\\vak_ry_dat.xlsx")


```
